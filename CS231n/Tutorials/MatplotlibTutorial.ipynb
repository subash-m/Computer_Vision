```python
import numpy as np
import matplotlib.pyplot as plt
from __future__ import print_function
```

## Compute the x and y coordinates for points on sine and cosine curves

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

## subplot(No_Rows, No_Cols, Index)

### In the current figure, create and return an Axes, at position index of a (virtual) grid of nrows by ncols axes. Indexes go from 1 to nrows * ncols, incrementing in row-major order.

```python
names = ['a', 'b', 'c']
values = [1, 3, 10]

plt.subplot(1, 3, 1)  # Shortform ----->  plt.subplot(131)
plt.bar(names, values)
```

## Make the Sine plot

```python
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title('Sine')
plt.show()
```

## Make the cosine plot.

```python
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')
plt.show()
```
